```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basics of Matlab

#Line plot
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x , y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label="Sine wave")
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend() #represents what color determines what value.
plt.show()
```

```python
#Bar Chart
categories = ['A', 'B', 'C']
values = [10, 20, 30]
plt.bar(categories, values, facecolor=["red", "green", "blue"], edgecolor="black", linewidth=2)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend()
plt.show()
```

#Visualization with Seaborn

```python
import seaborn as sns
```

```python
#Histogram and KDE plot:
data = np.random.randn(1000)
sns.histplot(data, kde=True, color="black")
plt.title("Histogram with KDE")
plt.show()
```

```python
#heatmap
matrix=np.random.rand(5,5)
sns.heatmap(matrix, annot=True, cmap="Greys", linecolor="white", linewidths=1.7)
plt.title("HITMAP EXAMPLE")
plt.show()
```

```python
#pairplot
sns.set_style("white")
df=sns.load_dataset("iris")
sns.pairplot(df, hue="species", markers=["o","s", "D"], palette=["red","green","purple"] )#square="s", circle="o", diamond="D"
plt.show()
```
